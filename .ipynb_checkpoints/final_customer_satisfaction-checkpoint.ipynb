# 4. Data Preprocessing

```python
# Clean and fill `Ticket Description` text.
TEXT_COL = 'Ticket Description'
TARGET_COL = 'Customer Satisfaction Rating'

# Ensure text exists
df[TEXT_COL] = df.get(TEXT_COL, '').fillna('').astype(str)



# Fill target missing with median and convert to int

df[TARGET_COL] = df[TARGET_COL].fillna(df[TARGET_COL].median()).astype(int)


# Derived text features- Create `desc_len` and `word_count`

df['desc_len'] = df[TEXT_COL].apply(lambda x: len(str(x)))
df['word_count'] = df[TEXT_COL].apply(lambda x: len(str(x).split()))



# Map continuous satisfaction ratings to 3 classes (unsatisfied=0, neutral=1, satisfied=2)
def map_rating(x):
    if x <= 2:
        return 0
    elif x == 3:
        return 1
    else:
        return 2

df['Satisfaction_Class'] = df[TARGET_COL].apply(map_rating)
print('3-class distribution:')
print(df['Satisfaction_Class'].value_counts(normalize=True))
```

```python
# Feature columns
cat_cols = [c for c in ['Customer Gender','Product Purchased','Ticket Type','Ticket Status','Ticket Priority','Ticket Channel'] if c in df.columns]
num_cols = [c for c in ['Customer Age','desc_len','word_count'] if c in df.columns]
print('Categorical cols:', cat_cols)
print('Numeric cols:', num_cols)
```

# 5. Feature Engineering function

```python
# TEXT, TF-IDF (1-3 grams) and feature selection
TFIDF_MAX = 15000
SELECT_K = 12000

tfidf = TfidfVectorizer(max_features=TFIDF_MAX, ngram_range=(1,3), stop_words='english', min_df=2)
X_text = tfidf.fit_transform(df[TEXT_COL].astype(str))
print('TF-IDF shape:', X_text.shape)

# Chi2 selection requires non-negative features (TF-IDF is non-negative)

SELECT_K = min(SELECT_K, X_text.shape[1])
selector = SelectKBest(chi2, k= SELECT_K)
X_text_sel = selector.fit_transform(X_text, df['Satisfaction_Class'])
print('Selected TF-IDF shape:', X_text_sel.shape)
```

```python
# One-hot encode categorical features and scale numeric features
from scipy import sparse

if len(cat_cols) > 0:
    ohe = OneHotEncoder(handle_unknown='ignore')
    X_cat = ohe.fit_transform(df[cat_cols])
    print('Categorical matrix shape:', X_cat.shape)
else:
    X_cat = sparse.csr_matrix(np.empty((df.shape[0],0)))

if len(num_cols) > 0:
    scaler = StandardScaler()
    X_num = scaler.fit_transform(df[num_cols])
    X_num = sparse.csr_matrix(X_num)
    print('Numeric matrix shape:', X_num.shape)
else:
    X_num = sparse.csr_matrix(np.empty((df.shape[0],0)))

# Combine features (sparse)
X = hstack([X_text_sel, X_cat, X_num], format='csr')
y = df['Satisfaction_Class']
print('Combined X shape:', X.shape)
```

# 6. Train Test Split

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print('Train:', X_train.shape, 'Test:', X_test.shape)
```

# 7. Modeling

```python
results = {}

# Random Forest (balanced)
rf = RandomForestClassifier(class_weight='balanced', n_estimators=300, min_samples_split=10, min_samples_leaf=2, random_state=42, max_features= 'sqrt', max_depth=20, n_jobs=-1)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)
results['RandomForest (weighted)'] = [accuracy_score(y_test, pred_rf), f1_score(y_test, pred_rf, average='macro')]


# XGBoost (default weighted behavior; multiclass scale_pos_weight not directly used)
xgb = XGBClassifier(objective='multi:softprob', num_class=3, learning_rate=0.01, max_depth=6, subsample=0.85,
                    colsample_bytree=0.8, n_estimators=500, eval_metric='mlogloss', use_label_encoder=False, n_jobs=-1)
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
results['XGBoost (default)'] = [accuracy_score(y_test, pred_xgb), f1_score(y_test, pred_xgb, average='macro')]

 
# Linear SVM (balanced)
svm = LinearSVC(class_weight='balanced', C= 2.0, loss="hinge", max_iter=10000)
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)
results['SVM (balanced)'] = [accuracy_score(y_test, pred_svm), f1_score(y_test, pred_svm, average='macro')]


pd.DataFrame(results, index=['Accuracy','Macro F1']).T
```

# 7. Model Evaluation

```python
# Final evaluation with bestTrain SVM (Final Predictor)

pred_svm = svm.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, pred_svm))
print("SVM Macro F1:", f1_score(y_test, pred_svm, average="macro"))
print(classification_report(y_test, pred_svm))

cm = confusion_matrix(y_test, pred_svm)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title('Final SVM Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import pandas as pd
import numpy as np

seg_df["Date of Purchase"] = pd.to_datetime(seg_df["Date of Purchase"], errors="coerce")
seg_df["First Response Time"] = pd.to_datetime(seg_df["First Response Time"], errors="coerce")
seg_df["Time to Resolution"] = pd.to_datetime(seg_df["Time to Resolution"], errors="coerce")


# Response time in hours
seg_df["first_response_hours"] = (
    seg_df["First Response Time"] - seg_df["Date of Purchase"]
).dt.total_seconds() / 3600

seg_df["first_response_hours"]= seg_df["first_response_hours"].astype("datetime64[ns]").view("int64")

# Resolution time in hours
seg_df["resolution_hours"] = (
    seg_df["Time to Resolution"] - seg_df["First Response Time"]
).dt.total_seconds() / 3600

seg_df["resolution_hours"]= seg_df["resolution_hours"].astype("datetime64[ns]").view("int64")
```

```python
# Create meaningful customer metrics

seg_df_grouped = seg_df.groupby("Customer Email").agg({
    "Customer Age": "mean",
    "Customer Gender": "first",
    "Product Purchased": lambda x: x.mode()[0] if len(x.mode()) else "Unknown",
    "Ticket Priority": lambda x: x.mode()[0] if len(x.mode()) else "Unknown",
    "Ticket Channel": lambda x: x.mode()[0] if len(x.mode()) else "Unknown",
    "Customer Satisfaction Rating": "mean",
    "resolution_hours": "mean",
    "first_response_hours": "mean",
    "Ticket Description": lambda x: x.apply(len).mean(),
}).reset_index()

seg_df_grouped.rename(columns={"Ticket Description": "avg_desc_len"}, inplace=True)

```

```python
seg_df_grouped.head()
```

```python
seg_df_grouped.info()
```

```python
seg_df_grouped = seg_df_grouped.drop(columns=["resolution_hours", "first_response_hours"])
```

# Encode + Scale

```python
# ----------------------------
# Encode categorical columns
# ----------------------------
cat_cols = ["Customer Gender", "Product Purchased", "Ticket Priority", "Ticket Channel"]

encoders = {}
for c in cat_cols:
    le = LabelEncoder()
    seg_df_grouped[c] = le.fit_transform(seg_df_grouped[c].astype(str))
    encoders[c] = le
```

```python
# ----------------------------
# Identify numeric columns ONLY
# ----------------------------
num_cols = []

for c in seg_df_grouped.columns:
    if c in ["Customer Email"] + cat_cols:
        continue
    if np.issubdtype(seg_df_grouped[c].dtype, np.number):
        num_cols.append(c)

print("Final numeric columns:", num_cols)
```

```python
seg_df_grouped[num_cols] = seg_df_grouped[num_cols].fillna(seg_df_grouped[num_cols].median())
```

```python
# ----------------------------
# Convert numeric columns to float
# ----------------------------
seg_df_grouped[num_cols] = seg_df_grouped[num_cols].astype(float)

# ----------------------------
# Scale numeric features
# ----------------------------
scaler_seg = StandardScaler()
X_seg = scaler_seg.fit_transform(seg_df_grouped[num_cols])

print("Segmentation matrix shape:", X_seg.shape)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
seg_df_grouped["Segment"] = kmeans.fit_predict(X_seg)

seg_df_grouped.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_seg)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=seg_df_grouped["Segment"], cmap="tab10")
plt.title("Customer Segmentation (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# TF-IDF (1-3 grams) and feature selection
TFIDF_MAX = 5000
SELECT_K = 200

tfidf_reg = TfidfVectorizer(max_features=TFIDF_MAX, ngram_range=(1,3), stop_words='english', min_df=2)
X_text = tfidf_reg.fit_transform(reg_df['Ticket Description'].astype(str))
print('TF-IDF shape:', X_text.shape)

# Chi2 selection requires non-negative features (TF-IDF is non-negative)
SELECT_K = min(SELECT_K, X_text.shape[1])

selector_reg = SelectKBest(f_regression, k=SELECT_K)
X_text_sel = selector_reg.fit_transform(X_text, y_res)

print("Selected text features:", X_text_sel.shape)
```

```python
# One-hot encode categorical features and scale numeric features
from scipy import sparse

if len(cat_cols) > 0:
    ohe_reg = OneHotEncoder(handle_unknown='ignore')
    X_cat = ohe.fit_transform(reg_df[cat_cols])
    print('Categorical matrix shape:', X_cat.shape)
else:
    X_cat = sparse.csr_matrix(np.empty((df.shape[0],0)))

if len(num_cols) > 0:
    scaler_reg = StandardScaler()
    X_num = scaler.fit_transform(reg_df[num_cols])
    X_num = sparse.csr_matrix(X_num)
    print('Numeric matrix shape:', X_num.shape)
else:
    X_num = sparse.csr_matrix(np.empty((reg_df.shape[0],0)))

# Combine features (sparse)
X_reg = hstack([X_text_sel, X_cat, X_num], format='csr')
y_res = reg_df['Ticket Resolution Time'].astype(float)
print('Combined X shape:', X.shape)
```

```python
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_res, test_size=0.2, random_state=42)
```

# Train RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

rf_reg = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf_reg.fit(Xr_train, yr_train)

pred_reg = rf_reg.predict(Xr_test)

mae = mean_absolute_error(yr_test, pred_reg)
mse = mean_squared_error(yr_test, pred_reg)
rmse = np.sqrt(mse)

print("Resolution Time MAE:", mae)
print("Resolution Time MSE:", mse)
print("Resolution Time RMSE:", rmse)
```

# SHAP for Regression

```python
explainer_reg = shap.TreeExplainer(rf_reg, 
    feature_perturbation="interventional")
sample_n = min(200, Xr_test.shape[0])
idx = np.random.choice(Xr_test.shape[0], sample_n, replace=False)

X_reg_shap = Xr_test[idx].toarray().astype(np.float32)

shap_values_reg = explainer_reg.shap_values(X_reg_shap, check_additivity=False)
shap.summary_plot(shap_values_reg, X_reg_shap)
```
